```python
import pickle
import cv2
import pylab as plt

graph = pickle.load(open('media/output.pickle', 'rb'))
image = cv2.imread('media/output.tif')

fig, ax = plt.subplots(figsize=(13, 13))
ax.imshow(image)

for u, neighbours in graph.items():
    for v in neighbours:
        plt.plot([u[0], v[0]], [u[1], v[1]], 'r')
        plt.scatter([u[0], v[0]], [u[1], v[1]], color='r')
```
