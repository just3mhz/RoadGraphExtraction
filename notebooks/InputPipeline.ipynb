```python
import tensorflow as tf
```

```python
import cv2
import re
import os
import tqdm
from glob import glob

import numpy as np

images = []
masks = []

def get_images():
    for image in tqdm.tqdm(glob('../data/Vegas/*_8bit.tif')):
        match = re.fullmatch(r'\.\./data/Vegas/([a-zA-Z0-9_]*)_8bit\.tif', image)
        if not match:
            continue
        mask = f'../data/Vegas/{match[1]}_mask.tif'
        if os.path.exists(mask):
            yield cv2.resize(cv2.imread(image), (512, 512)), cv2.resize(cv2.imread(image), (512, 512))
```

```python
dataset = tf.data.Dataset.from_generator(get_images,
                                         output_signature=(
                                             tf.TensorSpec(shape=(512, 512, 3), dtype=tf.uint8),
                                             tf.TensorSpec(shape=(512, 512, 3), dtype=tf.uint8)
                                         ))
```

```python
for d in dataset.take(10):
    print(d)
```
